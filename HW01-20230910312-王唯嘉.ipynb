```python
# -*- coding: utf-8 -*-
# 2.1
import numpy as np

# ======================
# 1. 向量定义
# ======================
a = np.array([2, -1, 3])
b = np.array([1, 4, -2])

print("===== 向量点积 a·b =====")
dot_result = np.dot(a, b)
print(f"a · b = {dot_result}")

print("\n===== 向量 a 的 2-范数（Frobenius 范数） =====")
norm_result = np.linalg.norm(a)
print(f"||a|| = {norm_result:.4f}")

# ======================
# 2. 矩阵乘法 A × B
# ======================
A = np.array([
    [1, 0, 2],
    [-1, 3, 1]
])

B = np.array([
    [2, 1],
    [0, -1],
    [3, 2]
])

print("\n===== 矩阵 A =====")
print(A)

print("\n===== 矩阵 B =====")
print(B)

print("\n===== 矩阵乘积 AB =====")
AB = A @ B
print(AB)

print("\n===== 结果汇总 =====")
print(f"向量点积 a·b = {dot_result}")
print(f"向量 a 的范数 = {norm_result:.4f}")
print("矩阵 AB = ")
print(AB)
```

```python
#2.2
import numpy as np

# ======================
# 任务1：创建 3×4 标准正态分布随机矩阵 X
# ======================
# np.random.randn() 生成标准正态分布（均值0，方差1）的随机数
X = np.random.randn(3, 4)
print("===== 任务1：矩阵 X (3×4 标准正态分布) =====")
print(X)
print("X 的形状：", X.shape)

# ======================
# 任务2：创建 4×2 全1矩阵 Y
# ======================
# np.ones() 生成全1矩阵
Y = np.ones((4, 2))
print("\n===== 任务2：矩阵 Y (4×2 全1矩阵) =====")
print(Y)
print("Y 的形状：", Y.shape)

# ======================
# 任务3：计算矩阵乘法 Z = X × Y
# ======================
# 矩阵乘法用 @ 运算符 或 np.dot()，推荐 @ 更直观
Z = X @ Y
print("\n===== 任务3：矩阵乘法 Z = X × Y =====")
print(Z)
print("Z 的形状：", Z.shape)  # 结果应为 3×2

# ======================
# 任务4：输出指定元素
# ======================
# Python 索引从0开始！
# 第一行第二列交叉处：行索引0，列索引1
z_0_1 = Z[0, 1]
# 第2行所有元素：行索引1
z_row_1 = Z[1, :]

print("\n===== 任务4：指定元素输出 =====")
print("Z 的第1行第2列交叉处元素（索引[0,1]）：", z_0_1)
print("Z 的第2行所有元素（索引[1,:]）：", z_row_1)

# ======================
# 任务5：计算 Z 的 Frobenius 范数
# ======================
# np.linalg.norm() 默认计算 Frobenius 范数
frobenius_norm = np.linalg.norm(Z)
print("\n===== 任务5：Z 的 Frobenius 范数 =====")
print(f"Z 的 Frobenius 范数 = {frobenius_norm:.4f}")  # 保留4位小数，更美观

print("\n===== 最终结果汇总 =====")
print(f"Z[0,1] = {z_0_1:.4f}")
print(f"Z第2行 = {np.round(z_row_1, 4)}")
print(f"Z的F范数 = {frobenius_norm:.4f}")
```

```python
# 3.1 贝叶斯公式计算
# 定义已知概率
P_A = 0.001       # 患病率 P(A)
P_notA = 1 - P_A  # 未患病概率 P(¬A)
P_B_givenA = 0.99 # 灵敏度 P(B|A)
P_B_given_notA = 0.02 # 假阳性率 P(B|¬A)

# 全概率公式计算 P(B)
P_B = P_A * P_B_givenA + P_notA * P_B_given_notA

# 贝叶斯公式计算 P(A|B)
P_A_givenB = (P_A * P_B_givenA) / P_B

# 打印结果
print("===== 3.1 贝叶斯公式计算结果 =====")
print(f"全概率 P(B) = {P_B:.6f}")
print(f"检测阳性时真正患病的概率 P(A|B) = {P_A_givenB:.4f} (约 {P_A_givenB*100:.2f}%)")
```

```python
# ======================
# 4.1 理论计算题：导数与梯度
# ======================
import sympy as sp

# 1. 定义符号变量
w1, w2 = sp.symbols('w1 w2')
x1, x2, y = 2, 1, 3 # 定义常数

# 定义函数 z
z = (w1 * x1 + w2 * x2 - y)**2

print("原函数 z =")
sp.pprint(z)

# 2. 计算偏导数表达式
dz_dw1 = sp.diff(z, w1)
dz_dw2 = sp.diff(z, w2)

print("\n1. 偏导数表达式:")
print("∂z/∂w1 =")
sp.pprint(dz_dw1)
print("∂z/∂w2 =")
sp.pprint(dz_dw2)

# 3. 代入数值计算梯度 (w1=0.5, w2=1)
w1_val = 0.5
w2_val = 1

# 计算 z 的值
z_val = z.subs({w1: w1_val, w2: w2_val})

# 计算梯度值
grad1_val = dz_dw1.subs({w1: w1_val, w2: w2_val})
grad2_val = dz_dw2.subs({w1: w1_val, w2: w2_val})

print("\n2. 当 w1=0.5, w2=1 时的数值结果:")
print(f"z = {z_val}")
print(f"梯度 ∂z/∂w1 = {grad1_val}")
print(f"梯度 ∂z/∂w2 = {grad2_val}")
```

```python
# ======================
# 4.2 编程题：反向传播与计算图（修复版）
# ======================
import torch

# ======================
# 1. 定义计算流程 & 2. 前向计算
# ======================
# 给定输入和参数
x = 2.0
w1_val = 1.5
w2_val = 0.5

# 前向计算：a = x * w1, b = a + w2, L = b^2
a = x * w1_val
b = a + w2_val
L = b ** 2

print("===== 4.2 前向计算结果 =====")
print(f"a = x * w1 = {x} * {w1_val} = {a}")
print(f"b = a + w2 = {a} + {w2_val} = {b}")
print(f"L = b^2 = {b}^2 = {L}")

# ======================
# 3. 手动反向传播 (链式法则) & PyTorch 验证
# ======================
# -------- 方法A：PyTorch 自动微分 (验证基准) --------
w1 = torch.tensor(w1_val, requires_grad=True)
w2 = torch.tensor(w2_val, requires_grad=True)
x_t = torch.tensor(x)

a_t = x_t * w1
b_t = a_t + w2
L_t = b_t ** 2

# 反向传播
L_t.backward()

print("\n===== 4.2 反向传播结果 =====")
print(f"PyTorch 自动微分结果:")
print(f"dL/dw1 (grad_w1) = {w1.grad.item()}")
print(f"dL/dw2 (grad_w2) = {w2.grad.item()}")

# -------- 方法B：手动链式法则计算（修复：用真实的b） --------
# 直接用前向计算出的真实b值，不再硬编码
dL_dw1_manual = 2 * b * x
dL_dw2_manual = 2 * b

print(f"\n手动链式法则计算结果:")
print(f"dL/dw1 = 2*b*x = 2 * {b} * {x} = {dL_dw1_manual}")
print(f"dL/dw2 = 2*b*1 = 2 * {b} = {dL_dw2_manual}")

# 验证一致性
print(f"\n===== 结果验证 =====")
print(f"w1梯度是否一致: {abs(w1.grad.item() - dL_dw1_manual) < 1e-6}")
print(f"w2梯度是否一致: {abs(w2.grad.item() - dL_dw2_manual) < 1e-6}")
```

```python
# ======================
# 5.2 编程题：Softmax 回归手写数字分类
# ======================

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# ======================
# 1. 加载数据集
# ======================
digits = load_digits()
X = digits.data
y = digits.target.reshape(-1, 1)

print("数据形状 X:", X.shape)
print("标签形状 y:", y.shape)

# ======================
# 2. 预处理：One-hot 编码 + 划分数据集
# ======================
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_onehot, test_size=0.2, random_state=42
)

print("训练集 X:", X_train.shape)
print("测试集 X:", X_test.shape)

# ======================
# 3. Softmax 函数
# ======================
def softmax(z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# ======================
# 4. 交叉熵损失
# ======================
def cross_entropy(y_pred, y_true):
    eps = 1e-10
    return -np.sum(y_true * np.log(y_pred + eps)) / len(y_pred)

# ======================
# 5. 初始化参数
# ======================
n_features = X_train.shape[1]
n_classes = y_train.shape[1]

W = np.random.randn(n_features, n_classes) * 0.01
b = np.zeros((1, n_classes))

# ======================
# 6. 小批量梯度下降
# ======================
lr = 0.1
epochs = 50
batch_size = 32
loss_list = []

for epoch in range(epochs):
    idx = np.random.permutation(len(X_train))
    X_shuffle = X_train[idx]
    y_shuffle = y_train[idx]

    total_loss = 0

    for i in range(0, len(X_train), batch_size):
        X_batch = X_shuffle[i:i+batch_size]
        y_batch = y_shuffle[i:i+batch_size]

        # 前向
        z = X_batch @ W + b
        y_pred = softmax(z)
        loss = cross_entropy(y_pred, y_batch)
        total_loss += loss * len(X_batch)

        # 反向
        dz = y_pred - y_batch
        dW = X_batch.T @ dz / len(X_batch)
        db = np.sum(dz, axis=0, keepdims=True) / len(X_batch)

        # 更新
        W -= lr * dW
        b -= lr * db

    avg_loss = total_loss / len(X_train)
    loss_list.append(avg_loss)

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1:2d}, Loss = {avg_loss:.4f}")

# ======================
# 7. 测试集准确率
# ======================
z_test = X_test @ W + b
y_pred_test = softmax(z_test)

pred = np.argmax(y_pred_test, axis=1)
true = np.argmax(y_test, axis=1)

acc = np.mean(pred == true)
print(f"\n测试集准确率 = {acc:.4f} ({acc*100:.2f}%)")

# ======================
# 8. 画损失曲线
# ======================
plt.figure(figsize=(10,4))
plt.plot(loss_list)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

```python
# ======================
# 6.1 理论计算题：正态分布的最大似然估计
# ======================

print("===== 6.1 理论推导 =====")
print("\n1. 似然函数 L(μ, σ²):")
print("对于独立同分布的正态样本 x₁, x₂, ..., xₙ ~ N(μ, σ²)")
print("似然函数是各样本概率密度的乘积：")
print("L(μ, σ²) = Π [1/(√(2πσ²)) * exp(-(xᵢ-μ)²/(2σ²))]")
print("对数似然函数：")
print("ln L = -n/2 * ln(2π) - n/2 * ln(σ²) - (1/(2σ²)) * Σ(xᵢ-μ)²")

print("\n2. 证明 μ 的最大似然估计为样本均值 μ̂ = (1/n)Σxᵢ:")
print("对 ln L 关于 μ 求导并令导数为0：")
print("d(ln L)/dμ = (1/σ²) * Σ(xᵢ-μ) = 0")
print("解得：Σxᵢ - nμ = 0  =>  μ̂ = (1/n)Σxᵢ")

print("\n3. 证明 σ² 的最大似然估计为 σ̂² = (1/n)Σ(xᵢ-μ̂)²:")
print("将 μ̂ 代入对数似然，对 σ² 求导并令导数为0：")
print("d(ln L)/d(σ²) = -n/(2σ²) + (1/(2(σ²)²)) * Σ(xᵢ-μ̂)² = 0")
print("解得：-n/2 + (1/(2σ²)) * Σ(xᵢ-μ̂)² = 0")
print("=> σ̂² = (1/n)Σ(xᵢ-μ̂)²")
```

```python
# ======================
# 6.2 编程题：从零实现二元逻辑回归
# ======================

import numpy as np
import matplotlib.pyplot as plt

# ======================
# 1. 生成两类线性可分的数据
# ======================
np.random.seed(42)

# 类别 0：围绕 (0,0) 分布
n_samples = 200
mean0 = [0, 0]
cov0 = [[1, 0], [0, 1]]
X0 = np.random.multivariate_normal(mean0, cov0, n_samples)
y0 = np.zeros(n_samples)

# 类别 1：围绕 (3,3) 分布
mean1 = [3, 3]
cov1 = [[1, 0], [0, 1]]
X1 = np.random.multivariate_normal(mean1, cov1, n_samples)
y1 = np.ones(n_samples)

# 合并数据
X_train = np.vstack((X0, X1))
y_train = np.hstack((y0, y1))

print("训练数据形状:", X_train.shape)
print("标签形状:", y_train.shape)

# 可视化训练数据
plt.figure(figsize=(8, 6))
plt.scatter(X0[:, 0], X0[:, 1], label='Class 0', alpha=0.6)
plt.scatter(X1[:, 0], X1[:, 1], label='Class 1', alpha=0.6)
plt.title('Generated Training Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# ======================
# 2. 定义sigmoid函数和二元交叉熵损失
# ======================
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def binary_cross_entropy(y_pred, y_true):
    # 防止log(0)
    eps = 1e-10
    return -np.mean(y_true * np.log(y_pred + eps) + (1 - y_true) * np.log(1 - y_pred + eps))

# ======================
# 3. 使用梯度下降法学习参数 w, b
# ======================
# 初始化参数
n_features = X_train.shape[1]
w = np.zeros(n_features)
b = 0

# 超参数
learning_rate = 0.1
epochs = 1000

loss_history = []

for epoch in range(epochs):
    # 前向传播
    z = np.dot(X_train, w) + b
    y_pred = sigmoid(z)
    
    # 计算损失
    loss = binary_cross_entropy(y_pred, y_train)
    loss_history.append(loss)
    
    # 反向传播：计算梯度
    m = len(X_train)
    dw = (1/m) * np.dot(X_train.T, (y_pred - y_train))
    db = (1/m) * np.sum(y_pred - y_train)
    
    # 更新参数
    w -= learning_rate * dw
    b -= learning_rate * db
    
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch+1:4d}, Loss: {loss:.4f}")

# 绘制损失曲线
plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

# ======================
# 4. 绘制决策边界并计算测试集准确率
# ======================
# 生成测试集
n_test = 100
X0_test = np.random.multivariate_normal(mean0, cov0, n_test)
X1_test = np.random.multivariate_normal(mean1, cov1, n_test)
X_test = np.vstack((X0_test, X1_test))
y_test = np.hstack((np.zeros(n_test), np.ones(n_test)))

# 预测测试集
z_test = np.dot(X_test, w) + b
y_pred_test = sigmoid(z_test)
y_pred_label = (y_pred_test > 0.5).astype(int)

# 计算准确率
accuracy = np.mean(y_pred_label == y_test)
print(f"\n测试集准确率: {accuracy:.4f} ({accuracy*100:.2f}%)")

# 绘制决策边界
plt.figure(figsize=(8, 6))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=plt.cm.Spectral, alpha=0.6)

# 决策边界方程: w1*x1 + w2*x2 + b = 0 => x2 = (-w1*x1 - b)/w2
x1_min, x1_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
x1 = np.linspace(x1_min, x1_max, 100)
x2 = (-w[0] * x1 - b) / w[1]

plt.plot(x1, x2, 'r-', label='Decision Boundary')
plt.title('Test Data with Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
